```python
import pandas as pd
import os
from sqlalchemy import create_engine
import logging
import time # 'time' was missing in your original import statements

# Make sure the 'logs' directory exists
if not os.path.exists('logs'):
    os.makedirs('logs')

logging.basicConfig(
    filename="logs/ingestion_db.log",
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s",
    filemode="a"
)

# Make sure the 'data' directory exists and contains CSVs
if not os.path.exists('data'):
    logging.error("The 'data' directory does not exist. Please create it and add your CSV files.")
else:
    engine = create_engine('sqlite:///inventory.db')

    def ingest_db(df, table_name, engine):
        df.to_sql(
            table_name,
            con=engine,
            if_exists='replace',
            index=False,
            chunksize=50000
        )

    def load_raw_data():
        '''this function will load the CSVs as dataframe and ingest into db'''
        start = time.time()
        for file in os.listdir('data'):
            if '.csv' in file:
                df = pd.read_csv('data/' + file)
                logging.info(f'Ingesting {file} in db')
                ingest_db(df, file[:-4], engine)
        end = time.time()
        total_time = (end - start) / 60
        logging.info('-----------------Ingestion Complete-----------------')
        logging.info(f'\nTotal Time Taken: {total_time} minutes')

    # This block must be at the top level of the script
    if __name__ == '__main__':
        load_raw_data()
```
